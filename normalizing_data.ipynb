```python
import pandas as pd
```

```python
df = pd.DataFrame(
    [
        [25000,200,30,5],
        [18000,150,12,5],
        [9000,100,25,7],
        [40000,300,21,2]
    ],
)
```

```python
display(df)
```

```python
#visualiz the data for see the diff in the dataset
import matplotlib.pyplot as plt
df.plot(kind = 'bar')
```

```python
#use the Maximum absolute scaling for normaliz data
```

```python
df_copy = df.copy()
```

```python
df_copy
```

```python
for col in df_copy.columns:
  df_copy[col] = df_copy[col] / df_copy[col].abs().max()
```
